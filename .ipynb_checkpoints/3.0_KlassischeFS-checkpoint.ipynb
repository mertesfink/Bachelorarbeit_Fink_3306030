# 1. Univariate Feature Selection (Korrelationskoeffizient)

## Optimaler Threshold

### Lineare Korrelationszusammenh√§nge (Heatmap)

```python
#Vorverarbeitete Daten aus CSV auslesen
df = pd.read_csv("data/df_preprocessed_all.csv", sep = ",")

#Das Label nennen
label = df.columns.get_loc('FELT_LIFE_NET')

N = 10
correlations = df.corrwith(df['FELT_LIFE_NET'], method='pearson')
correlation_df = pd.DataFrame({'Correlation': correlations})
correlation_df = correlation_df.abs().sort_values(by='Correlation', ascending=False)
top_features = correlation_df.index[:N]
# Erstelle eine Heatmap der korrelierten Features
plt.figure(figsize=(10, 8))
sns.heatmap(df[top_features].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap der Korrelationen zwischen den Top {} Features und {}'.format(N, "FELT_LIFE_NET"))
plt.show()
```

### Threshold bestimmen

```python
#Vorverarbeitete Daten aus CSV auslesen
df = pd.read_csv("data/df_preprocessed_all.csv", sep = ",")

#Das Label nennen
label = df.columns.get_loc('FELT_LIFE_NET')
header = df.columns

#Machine Learning Algorithms die benutzt werden
MLA = [
    ('Linear Regressor',LinearRegression()),
    ('SVR', SVR(kernel='linear')),
    ('Random Forest',RandomForestRegressor()),
    ('Gradient Boosting', xgb.XGBRegressor())
]

results = {}
selected_features = {}

# Farben f√ºr die Modelle definieren
colors = ['blue', 'green', 'red', 'black']

plt.figure(figsize=(10, 8))
plt.ylim(20, 85) #Y-Limit setzen, da Linearer Regressor und SVR anfangs sehr hohe Metrikwerte haben

for idx, (model_name, model_instance) in enumerate(MLA):
    
    
    threshold_values = []
    test_rmse_values = []
    test_mae_values = []
    num_features_values = []
    
    threshold = 0.0
    best_threshold_rmse = None
    best_threshold_mae = None
    min_rmse = float('inf')
    min_mae = float('inf')
    best_num_features_rmse = None
    best_num_features_mae = None
    
    
    while threshold <= 1:
        
        #Features f√ºr jeden Threshold berechnen und die dazugeh√∂rigen Metriken berechnen
        
        dataset = Dataset(df, 'df', label, divide_dataset=False, header=header)
        dataset.divide_dataset(model_instance, all_features=True, all_instances=True, evaluate=False, folds=5)

        selected_feature_names, num_selected_features, duration = featureselection.correlation(dataset.get_X_train(), dataset.get_y_train(), threshold=threshold)
        dataset.set_features(selected_feature_names)
        
        if(len(selected_feature_names)!=0):
           
            if(model_name =="Linear Regressor"):
                selected_features[threshold] = df.columns[selected_feature_names].tolist()

            dataset.fit_classifier()

            dataset.set_test_metrics()
            test = dataset.get_test_metrics()

            threshold_values.append(threshold)
            test_rmse_values.append(test['TestRMSE'])
            test_mae_values.append(test['TestMAE'])
            num_features_values.append(num_selected_features)

            # √úberpr√ºfe, ob aktueller Threshold der beste f√ºr RMSE ist
            if test['TestRMSE'] < min_rmse:
                min_rmse = test['TestRMSE']
                best_threshold_rmse = threshold
                best_num_features_rmse = num_selected_features

            # √úberpr√ºfe, ob aktueller Threshold der beste f√ºr MAE ist
            if test['TestMAE'] < min_mae:
                min_mae = test['TestMAE']
                best_threshold_mae = threshold
                best_num_features_mae = num_selected_features
           
            threshold += 0.01
           
        else:
            threshold = 2
        

    results[model_name] = {
        'Threshold': threshold_values,
        'TestRMSE': test_rmse_values,
        'TestMAE': test_mae_values,
        'Feature-Anzahl': num_features_values
    }
    
    # Plot f√ºr RMSE erstellen
    plt.plot(threshold_values, test_rmse_values, label=model_name + ' (Best Threshold RMSE: {:.2f})'.format(best_threshold_rmse), color=colors[idx])

    # Plot f√ºr MAE erstellen
    plt.plot(threshold_values, test_mae_values, linestyle='--', label=model_name + ' (Best Threshold MAE: {:.2f})'.format(best_threshold_mae), color=colors[idx])
    
    # Vertikale rote Linien hinzuf√ºgen, wenn sich die Feature-Anzahl √§ndert
    prev_num_features = None
    for i, num_features in enumerate(num_features_values):
        if prev_num_features is not None and num_features != prev_num_features:
            plt.axvline(x=threshold_values[i], color='grey', linestyle='-', linewidth=0.4)
            plt.text(threshold_values[i], plt.gca().get_ylim()[1] / 2, str(num_features), rotation=90, verticalalignment='center', horizontalalignment='right', fontsize=8)
        prev_num_features = num_features
    

plt.xlabel('Threshold')
plt.ylabel('Test Score')
plt.title('Test Score f√ºr verschiedene Modelle bei unterschiedlichen Threshold-Werten')
plt.legend(loc='upper right',prop={'size': 6}) # Legende in der oberen rechten Ecke

plt.grid(True)
plt.show()
```

## Vorhersage mit bestem Threshold

### Code

```python
#Vorverarbeitete Daten aus CSV auslesen
df = pd.read_csv("data/df_preprocessed_all.csv", sep = ",")

#Das Label nennen
label = df.columns.get_loc('FELT_LIFE_NET')
header = df.columns

#Machine Learning Algorithms die benutzt werden
MLA = [
    ('Linear Regressor',LinearRegression()),
    ('SVR', SVR(kernel='linear')),
    ('Random Forest',RandomForestRegressor()),
    ('Gradient Boosting', xgb.XGBRegressor())
]

results_corr = {}
predictions_corr = {}


# F√ºr alle Modelle in MLA FS mit Korrelationskeoffizienten mit anschlie√üender Validierung
for model_name, model_instance in MLA:
    
    dataset = Dataset(df, 'df', label, divide_dataset=False, header=header)
    dataset.divide_dataset(model_instance, ,normalize=False, all_features=True, all_instances=True, evaluate=False, folds=5)
    
    #FS mit Korrelationskoeffizient
    selected_features, num_selected_features, duration = featureselection.correlation(dataset.get_X_train(), dataset.get_y_train(), threshold=0.03)
    dataset.set_features(selected_features)
    
    #Regressor trainieren
    dataset.fit_classifier()
    
    # Cross-Validation
    dataset.set_CV()
    cv = dataset.get_CV()
    
    #Validierung auf Validation- und Testset
    dataset.set_train_metrics()
    train = dataset.get_train_metrics()
    traintime = dataset.get_traintime()
    
    dataset.set_validation_metrics()
    validation = dataset.get_validation_metrics()
    
    dataset.set_test_metrics()
    test = dataset.get_test_metrics()
    
    
    
    #dataset.plot_shapley_values("FS_corr_"+model_name)
    
    results_corr[model_name] = {
        'Model_name': model_name,
        'CV_TrainMAE': cv['CV_TrainMAE'],
        'CV_TrainRMSE': cv['CV_TrainRMSE'],
        'CV_TestMAE': cv['CV_TestMAE'],
        'CV_TestRMSE': cv['CV_TestRMSE'],
        'CV_fit_time': cv['CV_fit_time'],
        'CV_fit_time_ges':cv['CV_fit_time']+duration,
        'TrainRMSE': train['TrainRMSE'],
        'TrainMAE': train['TrainMAE'],
        'ValidationRMSE': validation['ValidationRMSE'],
        'ValidationMAE': validation['ValidationMAE'],
        'TestRMSE': test['TestRMSE'],
        'TestMAE': test['TestMAE'],
        'TrainTime': traintime,
        'TrainTime_ges': traintime+duration,
        'Features': selected_features,
        'Feature-Anzahl': num_selected_features,
        'FS-Laufzeit': duration
    }
    
    predictions_corr[model_name] = {
    'Model_name': model_name,
    'y_train': dataset.get_y_train(),
    'y_val': dataset.get_y_val(),
    'y_test': dataset.get_y_test(),
    'pred_train': dataset.get_y_pred_train(),
    'pred_val': dataset.get_y_pred_val(),
    'pred_test':dataset.get_y_pred_test()
    }
    print(model_name)
```

### Speichern der Ergebnisse (Corr)

```python
#Metriken speichern
df_corr = pd.DataFrame(results_corr.copy())
df_corr = df_corr.transpose()
df_corr.to_csv('data/Corr/All_FS_corr_metrics.csv', index=False)

#Ausgew√§hlten Features f√ºr die einzelnen Algorithmen speichern 
df_features_corr = pd.DataFrame()
# Iteration √ºber die Modelle und Hinzuf√ºgen der Feature-Namen in df_features_corr
for model_name, model_instance in MLA:
    selected_features = df.columns[results_corr[model_name]['Features']].tolist()
    
    # Die ausgew√§hlten Spaltennamen in eine neue Spalte in df_features_corr einf√ºgen
    df_features_corr[model_name] = selected_features

df_features_corr.to_csv('data/Corr/All_FS_corr_Features.csv', index=False)


#Predictions speichern
corr_data = []
# Iteriere √ºber die Modelle und ihre Daten
for model_name, data in predictions_corr.items():
    model_data = data.copy()
    model_data.pop('Model_name')  # Entferne den Eintrag 'Model_name'
    for data_type, values in model_data.items():
        # Iteriere √ºber die Werte in jedem Datenfeld und f√ºge sie zur flattened_data-Liste hinzu
        for value in values:
            corr_data.append({'Model_name': model_name, 'Data_type': data_type, 'Value': value})


df = pd.DataFrame(corr_data)
df.to_csv('data/Corr/All_FS_corr_predictions.csv', index=False)
```
